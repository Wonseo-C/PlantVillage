```python
# dataset analysis
# Download in https://data.mendeley.com/datasets/tywbtsjrjv/1
import enum
import os

dataDir = "./Plant_leave_diseases_dataset_with_augmentation/"
subDirList = os.listdir(dataDir)
subDirList = [subDir for subDir in os.listdir(dataDir) if not subDir.startswith('.')]

print("---The Plant Village's data category---")
total_dataNum = 0
for idx, category in enumerate(subDirList):
    print(f"category {idx+1}: {category}")
    dataNum = len(os.listdir(os.path.join(dataDir, category)))
    print(f"Data Num: {dataNum}")
    total_dataNum += dataNum
print(f"Total Data Num: {total_dataNum}")
```
